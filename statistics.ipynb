```python

# matplotlib.rc('font', family='Microsoft JhengHei')
matplotlib.rc('font', family='jf-openhuninn-1.1')


""" Import data """
df = pd.read_csv('voting.csv')

""" Print Columns """
row_cnt = df.shape[0]

new_df = df[["學號", "投下你神聖的一票", "第二票"]].drop_duplicates(
    subset="學號", keep="last")

print("remove duplicate rows: ", row_cnt - new_df.shape[0])
print(new_df.columns)

""" collecting all votes """

first_choice = new_df["投下你神聖的一票"].values
second_choice = new_df["第二票"].values

choices = np.concatenate((first_choice, second_choice), axis=0)

print("總票數: ", len(choices))

tag, cnt = np.unique(choices, return_counts=True)

plt.title("2023 工工系學會紀念品投票結果")
plt.pie(cnt, labels=tag, autopct='%1.1f%%')
plt.plot()
```
